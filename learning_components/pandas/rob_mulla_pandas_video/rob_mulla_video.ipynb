## FIltering Rows

```python
# df.iloc[] # Index location
# df.loc[] # Location using the name
df.iloc[1,3]
```

```python
df.iloc[5]
```

```python
df.iloc[[5]]
```

```python
df.iloc[:, 1] # all rows, first column
```

```python
df.loc[(df['Airline'] == 'Spirit Air Lines') & (df['FlightDate'] == '2022-08-09')]
```

```python
# use a tilda ~ to get the inverse of the boolean expression:
df.loc[~(df['Airline'] == 'Spirit Air Lines') & (df['FlightDate'] == '2022-08-09')]
```
